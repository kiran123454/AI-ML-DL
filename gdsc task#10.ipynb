```python
import pandas as pd
import numpy as np

# Create sample wine review data
data = {
    'country': ['Australia', 'New Zealand', 'France', 'Australia', 'Italy', 
               'New Zealand', 'Australia', 'US', 'New Zealand', 'Spain'],
    'points': [96, 95, 90, 97, 92, 96, 95, 94, 97, 91],
    'price': [65.0, 45.0, 78.0, 89.0, 55.0, 62.0, 58.0, 47.0, 72.0, 38.0],
    'province': ['Victoria', 'Marlborough', 'Bordeaux', 'South Australia',
                'Tuscany', 'Marlborough', 'Western Australia', 'California',
                'Central Otago', 'Rioja'],
    'variety': ['Shiraz', 'Sauvignon Blanc', 'Merlot', 'Cabernet Sauvignon',
               'Sangiovese', 'Pinot Noir', 'Chardonnay', 'Zinfandel',
               'Pinot Noir', 'Tempranillo']
}

reviews = pd.DataFrame(data)

# Filter for top Oceania wines (95+ points from Australia or New Zealand)
top_oceania_wines = reviews[
    (reviews['points'] >= 95) & 
    (reviews['country'].isin(['Australia', 'New Zealand']))
].copy()

# Display results
print("Top Oceania Wines (95+ points):")
print(f"Found {len(top_oceania_wines)} matching wines")
print("\nCountries represented:", top_oceania_wines['country'].unique())
print("Points range:", top_oceania_wines['points'].min(), "to", top_oceania_wines['points'].max())
print("\nSample results:")
display(top_oceania_wines.head())
```

```python
import pandas as pd

# Create sample data (replace with your actual data if needed)
data = {
    'title': ['Château Lafite', 'Two Buck Chuck', 'Vega Sicilia', 'Yellow Tail'],
    'points': [95, 85, 98, 88],
    'price': [900, 5, 350, 12]
}
reviews = pd.DataFrame(data)

# Calculate points-to-price ratio
reviews['value_ratio'] = reviews['points'] / reviews['price']

# Find the best bargain
bargain_idx = reviews['value_ratio'].idxmax()
bargain_wine = reviews.loc[bargain_idx, 'title']

# Display results
print("All wines with their value ratios:")
print(reviews[['title', 'points', 'price', 'value_ratio']].sort_values('value_ratio', ascending=False))
print(f"\nBest bargain wine: '{bargain_wine}'")
print(f"Its points-to-price ratio: {reviews.loc[bargain_idx, 'value_ratio']:.2f}")
```

```python
import pandas as pd

# Create sample data (replace with your actual data if needed)
data = {
    'description': [
        'This tropical wine has fruity notes of pineapple',
        'A fruity wine with berry flavors',
        'Tropical fruit aromas with a hint of citrus',
        'Rich and fruity with tropical undertones',
        'Smooth with no tropical or fruity characteristics'
    ]
}
reviews = pd.DataFrame(data)

# Count occurrences (case-insensitive)
tropical_count = reviews['description'].str.contains('tropical', case=False).sum()
fruity_count = reviews['description'].str.contains('fruity', case=False).sum()

# Create the Series
descriptor_counts = pd.Series({
    'tropical': tropical_count,
    'fruity': fruity_count
})

# Display results
print("Word occurrences in wine descriptions:")
print(descriptor_counts)

# Determine which is more common
more_common = 'fruity' if fruity_count > tropical_count else 'tropical'
print(f"\nThe word '{more_common}' appears more frequently in descriptions")
```

```python
import pandas as pd
import numpy as np

# Create sample data (replace with your actual data if needed)
data = {
    'points': [96, 88, 92, 84, 97, 82, 99],
    'country': ['Canada', 'US', 'France', 'Italy', 'Canada', 'Spain', 'Germany']
}
reviews = pd.DataFrame(data)

# Create star ratings based on conditions
conditions = [
    reviews['country'] == 'Canada',  # Canadian wines get 3 stars
    reviews['points'] >= 95,         # 95+ points gets 3 stars
    reviews['points'] >= 85,         # 85-94 points gets 2 stars
    True                             # All others get 1 star
]

choices = [3, 3, 2, 1]

star_ratings = pd.Series(
    np.select(conditions, choices),
    index=reviews.index,
    name='star_ratings'
)

# Display results
print("Wine reviews with star ratings:")
print(pd.concat([reviews, star_ratings], axis=1))
```

```python
import pandas as pd

# Create sample data (replace with your actual data if needed)
data = {
    'price': [15, 50, 100, 50, 200, 15, 100],
    'points': [88, 92, 95, 90, 96, 89, 94]
}
reviews = pd.DataFrame(data)

# Create Series of max points per price
best_wines = reviews.groupby('price')['points'].max().sort_index()

# Display results
print("Maximum points achieved at each price point:")
print(best_wines)

# Optional: Find the best value wines (highest points per dollar)
print("\nBest value wines (points per dollar):")
print((best_wines / best_wines.index).sort_values(ascending=False).head())
```

```python
import pandas as pd

# Create sample data (replace with your actual data if needed)
data = {
    'variety': ['Chardonnay', 'Pinot Noir', 'Cabernet', 'Chardonnay', 
               'Pinot Noir', 'Merlot', 'Cabernet', 'Merlot'],
    'price': [15, 25, 45, 20, 30, 22, 50, 18]
}
reviews = pd.DataFrame(data)

# Create DataFrame with min and max prices per variety
price_ranges = reviews.groupby('variety')['price'].agg(['min', 'max'])

# Display results
print("Price ranges for each wine variety:")
print(price_ranges)

# Optional: Add range column (max - min)
price_ranges['range'] = price_ranges['max'] - price_ranges['min']
print("\nWith price range calculation:")
print(price_ranges)
```

```python
import pandas as pd

# Using the previous example's output as starting point
price_ranges = pd.DataFrame({
    'min': [45, 15, 18, 25],
    'max': [50, 20, 22, 30],
    'range': [5, 5, 4, 5]
}, index=['Cabernet', 'Chardonnay', 'Merlot', 'Pinot Noir'])

# Sort varieties by min price (descending), then max price (descending)
sorted_varieties = price_ranges.sort_values(
    by=['min', 'max'],
    ascending=[False, False]
).copy()

# Display results
print("Wine varieties sorted by price (most expensive first):")
print(sorted_varieties)

# Optional: Highlight the top 3 most expensive varieties
print("\nTop 3 most expensive varieties:")
print(sorted_varieties.head(3))
```

```python
import pandas as pd

# Create sample data (replace with your actual data if needed)
data = {
    'taster_name': ['Jane Smith', 'Robert Parker', 'Jane Smith', 
                   'Wine Enthusiast', 'Robert Parker', 'James Suckling'],
    'points': [92, 95, 88, 90, 96, 93]
}
reviews = pd.DataFrame(data)

# Calculate average points per reviewer
avg_scores = reviews.groupby('taster_name')['points'].mean()

# Sort by average score (descending) and display
avg_scores = avg_scores.sort_values(ascending=False)
print("Average review scores by critic:")
print(avg_scores)

# Optional: Show count of reviews per critic
print("\nNumber of reviews per critic:")
print(reviews['taster_name'].value_counts())
```

```python
import pandas as pd

# Create sample data (replace with your actual data if needed)
data = {
    'country': ['US', 'France', 'Italy', 'US', 'France', 'Italy', 'US', 'Spain'],
    'variety': ['Pinot Noir', 'Chardonnay', 'Sangiovese', 'Pinot Noir', 
               'Merlot', 'Sangiovese', 'Cabernet', 'Tempranillo']
}
reviews = pd.DataFrame(data)

# Create MultiIndex Series sorted by frequency
country_variety_counts = reviews.groupby(['country', 'variety']).size() \
                              .sort_values(ascending=False) \
                              .rename('wine_count')

# Display results
print("Most common country-variety combinations:")
print(country_variety_counts)

# Optional: Show top 3 combinations
print("\nTop 3 combinations:")
print(country_variety_counts.head(3))
```
