```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load your local CSV
file_path = r"C:\Users\HP\fifa.csv"
df = pd.read_csv(file_path)

# 2. Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="pace",               # Speed attribute :contentReference[oaicite:4]{index=4}
    y="shooting",           # Shooting skill :contentReference[oaicite:5]{index=5}
    hue="nationality",      # Color points by nationality :contentReference[oaicite:6]{index=6}
    alpha=0.7,              # Slight transparency to reduce overplotting :contentReference[oaicite:7]{index=7}
    palette="tab10",        # A qualitative palette good for up to 10 categories :contentReference[oaicite:8]{index=8}
    edgecolor=None
)
plt.title("FIFA Players: Pace vs. Shooting by Nationality")
plt.xlabel("Pace")
plt.ylabel("Shooting")
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left", title="Nationality")
plt.tight_layout()
plt.show()
```
